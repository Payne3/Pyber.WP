```python
%matplotlib inline
%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_ = "data/city_data.csv"
ride_data_= "data/ride_data.csv"

# Read the City and Ride Data
city_data_df= pd.read_csv(city_data_)
ride_data_df = pd.read_csv(ride_data_)
# Combine the data into a single dataset
city_ride_df = pd.merge(city_data_df, ride_data_df, on="city",how="inner")
# Display the data table for preview
city_ride_df.head()
```

```python

#calculate urban metrics we are meausring 
urban_data = city_ride_df.loc[city_ride_df["type"]=="Urban"]
urban_rider_cnt = urban_data.groupby(["city"]).count()["ride_id"]
urban_fare_avg = urban_data.groupby(["city"]).mean()["fare"]
urban_driver_cnt = city_data_df.loc[city_data_df["type"]=="Urban"]
urban_driver = urban_driver_cnt["driver_count"]

#calculate suburban metrics
suburban_data = city_ride_df.loc[city_ride_df["type"]=="Suburban"]
suburban_rider_cnt = suburban_data.groupby(["city"]).count()["ride_id"]
suburban_fare_avg = suburban_data.groupby(["city"]).mean()["fare"]
suburban_driver_cnt = city_data_df.loc[city_data_df["type"]=="Suburban"]
suburban_driver = suburban_driver_cnt["driver_count"]

#calculate Rural metrics
rural_data = city_ride_df.loc[city_ride_df["type"]=="Rural"]
rural_rider_cnt = rural_data.groupby(["city"]).count()["ride_id"]
rural_fare_avg = rural_data.groupby(["city"]).mean()["fare"]
rural_driver_cnt = city_data_df.loc[city_data_df["type"]=="Rural"]
rural_driver = rural_driver_cnt["driver_count"]
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
plt.plot
plt.xlim(0,40)
plt.ylim(18,45)
plt.title("Urban Rider Count by Avg Fare")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Avg Fare $")
plt.grid()



# plt.legend(labels)
Big_Urban = plt.scatter(urban_rider_cnt, urban_fare_avg ,marker="o", s=urban_driver*12, facecolors="red", edgecolors="black", alpha=0.4, label= "Urban")
Big_Suburban = plt.scatter(suburban_rider_cnt, suburban_fare_avg ,marker="o", s=suburban_driver*12, facecolors="blue", edgecolors="black", alpha=0.4, label="Suburban")
Big_Rural = plt.scatter(rural_rider_cnt, rural_fare_avg ,marker="o", s=rural_driver*12, facecolors="green", edgecolors="black", alpha=0.4, label="Rural")

plt.legend(handles=[Big_Urban, Big_Suburban ,Big_Rural], loc="best")
plt.text(40,30, "*Cirle size correlates with Driver count")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

urban_fare_total = urban_data.groupby(["city"]).sum()["fare"]
suburban_fare_total = suburban_data.groupby(["city"]).sum()["fare"]
rural_fare_total = rural_data.groupby(["city"]).sum()["fare"]




urban = urban_fare_total.sum()
suburban = suburban_fare_total.sum()
rural = rural_fare_total.sum()


pie_data = [urban, suburban, rural]

colors = ["red", "coral", "yellow"]

labels = ["Urban", "Suburban", "Rural"]

explode = (0.1,0,0)

plt.pie(pie_data, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
plt.title("Total Fares by City Type")

# Build Pie Chart
plt.show()

# Save Figure
plt.savefig("../Images/PyPies.png")
```
